````python
def parse_meal_info(json_str):
  """
  Parses a JSON string containing meal information and returns separate variables
  for each meal and the explanation.

  Args:
      json_str: The JSON string to parse.

  Returns:
      A dictionary containing separate variables for each meal ("Meal_1", "Meal_2", "Meal_3")
      and the explanation.
  """
  meal_1 = None
  meal_2 = None
  meal_3 = None
  explanation = None

  # Find starting and ending indices for "Meal_Recommendations"
  start_idx = json_str.find('"Meal_Recommendations"')
  end_idx = json_str.find("} \n}", start_idx) + 1

  # Extract "Meal_Recommendations" section
  meal_recommendations = json_str[start_idx:end_idx]

  # Find all meal names and corresponding values
  for meal_name in ["Meal_1", "Meal_2", "Meal_3"]:
    name_start = meal_recommendations.find(f'"{meal_name}"')
    if name_start != -1:
      # Extract value after colon
      value_start = meal_recommendations.find(":", name_start) + 1
      value_end = meal_recommendations.find(",", value_start)
      if value_end == -1:
        value_end = len(meal_recommendations)
      # Extract the complete value (including quotes) for meals
      meal_value = meal_recommendations[value_start:value_end].strip()
      if meal_name == "Meal_1":
        meal_1 = meal_value
      elif meal_name == "Meal_2":
        meal_2 = meal_value
      else:
        meal_3 = meal_value

  # Find explanation
  explanation_start = json_str.find('"Explanation"')
  if explanation_start != -1:
    value_start = json_str.find(":", explanation_start) + 1
    value_end = json_str.find(",", value_start)
    if value_end == -1:
      value_end = len(json_str)
    explanation = json_str[value_start:value_end].strip()

  return {
    "Meal_1": meal_1,
    "Meal_2": meal_2,
    "Meal_3": meal_3,
    "Explanation": explanation
  }

# Example usage
json_data = """{ ... your JSON string here ... }"""
meal_info = parse_meal_info(response.text)

print(meal_info["Meal_1"])
print(meal_info["Meal_2"])
print(meal_info["Meal_3"])
print(meal_info["Explanation"])
````
